```python
# Univariate Analysis

plt.figure(figsize= (10, 6))
plt.hist(tips["total_bill"])
plt.title("Histogram of total_bill")
plt.xlabel("size")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure(figsize= (10, 6))
sns.boxplot(x = tips["total_bill"])
plt.title("Box plot of total_bill")
plt.xlabel("Total bill")
plt.show()
```

```python
tips[tips["total_bill"] > 50 ]
```

```python

plt.figure(figsize= (10, 6))
sns.kdeplot(tips["tip"], fill = True)
plt.title("Histogram/Density of  tip")
plt.xlabel("tip")
plt.show()
```

```python
tips.dtypes
```

```python

plt.figure(figsize= (10, 6))
sns.countplot(x = tips["day"])
plt.title("days of Weeks")
plt.xlabel("day")
plt.show()
```

```python

plt.figure(figsize= (10, 6))
tips['day'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.title("days of Weeks")
# plt.xlabel("day")
plt.ylabel("")
plt.show()
```

```python
# Bivariate Analysis - two variables at a time 
# Scatterplot

plt.figure(figsize=(10, 6))
sns.scatterplot(x = tips['total_bill'], y = tips['tip'])
plt.title("Scatterplot of total bill vs tip of the waiter")
plt.xlabel("Total bill")
plt.ylabel("tip of the waiter")
plt.show()
```

```python
#line plot

plt.figure(figsize=(10, 6))
sns.lineplot(x = tips['total_bill'], y = tips['tip'])
plt.title("lineplot of total bill vs tip of the waiter")
plt.xlabel("Total bill")
plt.ylabel("tip of the waiter")
plt.show()
```

```python
# barplot 

plt.figure(figsize=(10, 6))
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Barplot of total bill vs Weeks days")
plt.xlabel("Week days")
plt.ylabel("total_bill")
plt.show()
```

```python
# Boxplot 

plt.figure(figsize=(10, 6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Boxplot of total bill vs Weeks days")
plt.xlabel("Week days")
plt.ylabel("total_bill")
plt.show()
```
